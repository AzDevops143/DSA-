```python
import math

def detailed_merge_with_trace(left, right):
    result = []
    i = j = split_inv = 0
    print(f"\n      --- Starting Merge Detail ---")
    print(f"      Left: {left} | Right: {right}")

    # Step-by-step arrangement logic
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            print(f"      Compare {left[i]} and {right[j]}: {left[i]} is smaller. Adding to list.")
            result.append(left[i])
            i += 1
        else:
            print(f"      Compare {left[i]} and {right[j]}: {right[j]} is smaller! Inversions found: {len(left)-i}")
            result.append(right[j])
            split_inv += (len(left) - i)
            j += 1
        print(f"      Current List: {result}")

    # Appending remaining elements
    if i < len(left):
        print(f"      Appending remaining from Left: {left[i:]}")
        result.extend(left[i:])
    if j < len(right):
        print(f"      Appending remaining from Right: {right[j:]}")
        result.extend(right[j:])

    print(f"      Final Merged Result: {result}")
    return result, split_inv

def count_inversions_with_trace(arr, depth=0):
    indent = "    " * depth
    n = len(arr)

    if n <= 1:
        return arr, 0

    mid = n // 2
    print(f"{indent}\u25b6 DIVIDE: {arr} into {arr[:mid]} and {arr[mid:]}")

    left_sorted, left_inv = count_inversions_with_trace(arr[:mid], depth + 1)
    right_sorted, right_inv = count_inversions_with_trace(arr[mid:], depth + 1)

    # Use detailed trace for the final root merge
    if depth == 0:
        print(f"\n{'='*20} FINAL STAGE LOGIC {'='*20}")
        merged, split_inv = detailed_merge_with_trace(left_sorted, right_sorted)
    else:
        # Standard merge for internal levels
        merged = []
        i = j = split_inv = 0
        while i < len(left_sorted) and j < len(right_sorted):
            if left_sorted[i] <= right_sorted[j]:
                merged.append(left_sorted[i]); i += 1
            else:
                merged.append(right_sorted[j]); split_inv += (len(left_sorted)-i); j += 1
        merged.extend(left_sorted[i:]); merged.extend(right_sorted[j:])

    total_inv = left_inv + right_inv + split_inv
    print(f"{indent}\u25c0 MERGE: {left_sorted} + {right_sorted} \u2192 {merged}")
    print(f"{indent}  RESULT: Local Inversions = {total_inv}")

    return merged, total_inv

# --- Execution ---
input_arr = [1, 3, 5, 2, 4, 6, 9, 2, 7]
print(f"ASSIGNMENT: O(n log n) INVERSION COUNTING TRACE\n")
final_sorted, total = count_inversions_with_trace(input_arr)

print(f"\n--- FINAL SUBMISSION DATA ---")
print(f"Total Inversions: {total}")
print(f"Sorted Output: {final_sorted}")
```
