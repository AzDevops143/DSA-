```python
def count_inversions_and_sort(arr, level=0):
    # Base case: A single element has 0 inversions and is already sorted
    if len(arr) <= 1:
        return arr, 0

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    print(f"{'  ' * level}Splitting: {arr} into {left_half} and {right_half}")

    # Recursively find inversions in both halves
    sorted_left, left_inv = count_inversions_and_sort(left_half, level + 1)
    sorted_right, right_inv = count_inversions_and_sort(right_half, level + 1)

    # Merge the sorted halves and count split inversions
    merged_arr, split_inv = merge_and_count_split(sorted_left, sorted_right)

    total_inv = left_inv + right_inv + split_inv

    print(f"{'  ' * level}Merged stage: {merged_arr} | Split Inversions found: {split_inv}")

    return merged_arr, total_inv

def merge_and_count_split(left, right):
    result = []
    i = j = 0
    count = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            # If left[i] > right[j], then all remaining elements in 'left'
            # (from index i to the end) are greater than right[j].
            result.append(right[j])
            count += (len(left) - i)
            j += 1

    # Append remaining elements
    result.extend(left[i:])
    result.extend(right[j:])

    return result, count

# Initial Data from your image
A = [8, 4, 2, 1, 9, 3, 5, 6]

print("--- Starting Inversion Count Process ---")
final_sorted, total_count = count_inversions_and_sort(A)
print("--- Goal Stage Reached ---")
print(f"Final Sorted Array: {final_sorted}")
print(f"Total Inversion Count: {total_count}")
```

```python
def count_inversions_and_sort(arr, level=0):
    # Base case
    if len(arr) <= 1:
        return arr, 0

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # 1. DIVIDE
    # The 'level' helps us indent the output to show the tree structure
    indent = "    " * level
    print(f"{indent}▶ Dividing {arr}")

    sorted_left, left_inv = count_inversions_and_sort(left_half, level + 1)
    sorted_right, right_inv = count_inversions_and_sort(right_half, level + 1)

    # 2. CONQUER & MERGE
    merged_arr, split_inv = merge_and_count_split(sorted_left, sorted_right)

    # 3. CALCULATION VISIBILITY
    total_inv = left_inv + right_inv + split_inv

    print(f"{indent}◀ Merged: {merged_arr}")
    print(f"{indent}  CALCULATION: {left_inv} (Left) + {right_inv} (Right) + {split_inv} (Split) = {total_inv} Total\n")

    return merged_arr, total_inv

def merge_and_count_split(left, right):
    result = []
    i = j = 0
    count = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            # When an element from the right is smaller than the left,
            # it is an inversion relative to ALL remaining elements in the left.
            result.append(right[j])
            count += (len(left) - i)
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result, count

# Test Data
A = [8, 4, 2, 1, 9, 3, 5, 6]

print("--- INVERSION COUNTING START ---\n")
final_sorted, grand_total = count_inversions_and_sort(A)
print("--- GOAL STAGE REACHED ---")
print(f"Final Sorted Array: {final_sorted}")
print(f"Grand Total Inversions: {grand_total}")
```
